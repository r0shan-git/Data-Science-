```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
#Data
labels=["Sachine","Sehwag","Kohli","Yuraj"]
runs=[18000,8000,12000,9500]

plt.title("Career Runs of India Batsmen")
plt.pie(runs,labels=labels)
plt.show()
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

```python
plt.pie(runs,labels=labels,
        colors=colors,
        wedgeprops={'edgecolor':'#127806','linewidth':2,'linestyle': '--','fill':True})
plt.show()
```

```python
explode=[0.1,0,0,0]
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True,startangle=10 ,autopct='%1.1f%%' # adds a 3D-like shadow
)
plt.title("Exploded pie Examples")
plt.show()
```
