```python
class Employee:
    company="HP"
    def  __init__(self,name,salary):
        self.name=name
        self.salary=salary
    
    def printDetails(self):
        print(f"The company of {self.name} is {self.company}  with salary {self.salary}")
    
    @staticmethod
    def printTime():
        print(f"The time is now ")
        
e=Employee("Roshan",470000)
e.printDetails()
e.printTime()
```

```python
# if we want make classmethod
class Employee:
    company="HP"
    def  __init__(self,name,salary):
        self.name=name
        self.salary=salary
    
    def printDetails(self):
        print(f"The company of {self.name} is {self.company}  with salary {self.salary}")
    
    @staticmethod
    def printTime():
        print(f"The time is now ")
    def printClassDetails(cls):
        print(f"The company is {cls.company}")
        
e=Employee("Roshan",470000)
e.printDetails()
e.printTime()
e.printClassDetails()
```
